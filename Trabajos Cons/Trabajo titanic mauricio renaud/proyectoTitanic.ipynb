4.-***Graficamos el dataframe con Pandas**

Utilizo el .plot que es una funcion para graficar con pandas sin necesidad de usar matplotlib

```python
#Uso df.("Columna").value_counts() para contar cuantas veces aparece cada valor unico en una columna, devuelve una serie ordenada
df["Age"].value_counts()
```

```python
#Asigno varios colores a la variable colores para ver mis graficas de diferentes tonos 
colores = ['red', 'blue', 'green', 'orange', 'purple','pink','black']
```

Variable Numerica

```python
#Edad de los pasajeros de mas cantidad de pasajeros a menos

edad=df["Age"].value_counts().plot(kind="bar",color=colores, edgecolor="black",figsize=(20, 6))#asigno la columna age y ajusto el tipo de grafica, su color y tamaño.
edad.set_title("Edad de los pasajeros", fontsize=20, fontweight="bold", fontfamily="serif")#pongo titulo y ajusto parametros
edad.set_xlabel("Edad", fontsize=14, fontweight="bold", fontfamily="serif")
edad.set_ylabel("Numero de Pasajeros", fontsize=14, fontweight="bold", fontfamily="serif")
```

```python
#Edad de los pasajeros ordenada con (sort_index()) muestra de menor a mayor edad

edad=df["Age"].value_counts().sort_index().plot(kind="bar",color=colores, edgecolor="black",figsize=(20, 6))#asigno la columna age y ajusto el tipo de grafica, su color y tamaño.
edad.set_title("Edad de los pasajeros", fontsize=20, fontweight="bold", fontfamily="serif")#pongo titulo y ajusto parametros
edad.set_xlabel("Edad", fontsize=14, fontweight="bold", fontfamily="serif")
edad.set_ylabel("Numero de Pasajeros", fontsize=14, fontweight="bold", fontfamily="serif")
```

Variable Categorica

```python
#Grafica de barras del puerto por donde embarco cada pasajero 
edad=df["Embarked"].value_counts().plot(kind="bar",color="magenta", edgecolor="black")#asigno la columna age y ajusto el tipo de grafica y su color
edad.set_title("Cantidad de Pasajeros por Puerto de Embarque", fontsize=20, fontweight="bold", fontfamily="serif")#pongo titulo y ajusto parametros
edad.set_xlabel("Tipo de embarcacion", fontsize=14, fontweight="bold", fontfamily="serif")
edad.set_ylabel("Pasajeros", fontsize=14, fontweight="bold", fontfamily="serif")
```

### 5.-*Graficamos el dataframe con *matplotlib**

**Importamos matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
plt.figure(figsize=(20, 10))
plt.hist(df['Age'].dropna(),bins=30, color='purple',edgecolor="black",) 
plt.ylim()
plt.xlim(0,80)
plt.title("Edad", fontdict={"fontsize": 24, "fontweight": "bold"})
plt.xlabel("Edad",fontdict={"fontsize": 14, "fontstyle": "italic"})
plt.ylabel("cantidad de pasajeros", fontdict={"fontsize": 14, "fontstyle": "italic"})
plt.show()
```

```python
Edad= df["Age"].value_counts()
colores = ['red', 'blue', 'green', 'orange', 'purple']

plt.figure(figsize=(15, 10)) 
Edad.plot(kind="bar", color=colores, edgecolor="black")
plt.xlabel("Edad", fontsize=12, fontweight="bold")
plt.ylabel("Cantidad de pasajeros", fontsize=12, fontweight="bold")
plt.title("Edad de los pasajeros", fontsize=20, fontweight="bold")
```

```python
clase= df["Pclass"].value_counts()
plt.figure(figsize=(15, 10)) 
clase.plot(kind="pie", color="yellow", )
plt.xlabel("Clase", fontsize=12, fontweight="bold")
plt.ylabel("Cantidad de Personas", fontsize=12, fontweight="bold")
plt.title("Tipo de Clase", fontsize=20, fontweight="bold")
```

```python
Supervivientes= df["Survived"].value_counts()
plt.figure(figsize=(10, 6)) 
Supervivientes.plot(kind="bar", color="yellow", edgecolor="black")

plt.xlabel("Sobrevivieron 1 No Sobrevivieron 0", fontsize=12, fontweight="bold")
plt.ylabel("Cantidad de Personas", fontsize=12, fontweight="bold")
plt.title("Supervivientes", fontsize=14, fontweight="bold")

plt.show()
```
